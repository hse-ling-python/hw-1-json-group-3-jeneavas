```python
%load_ext pycodestyle_magic
%pycodestyle_on

import json
from pprint import pprint
from collections import Counter

twitter = []
for line in open('hw_3_twitter.json'):
    twitter.append(json.loads(line))
pprint(twitter[0])
```

# –í–æ–ø—Ä–æ—Å 1

–ö–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞ - —Ç–≤–∏—Ç. –ß—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–≤–∏—Ç–æ–≤ –ø—Ä–æ—Å—Ç–æ –ø–æ—Å—á–∏—Ç–∞–µ–º –¥–ª–∏–Ω—É —Å–ø–∏—Å–∫–∞.


```python
print('–≤ –Ω–∞–±–æ—Ä–µ', len(twitter), '—Ç–≤–∏—Ç–æ–≤')
```

# –í–æ–ø—Ä–æ—Å 2

–ü—Ä–æ–≤–µ—Ä–∏–º, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∫–ª—é—á —Ç–≤–∏—Ç–∞ —Å—Ç—Ä–æ–∫–æ–π 'deleted'. –ï—Å–ª–∏ –¥–∞, –±—É–¥–µ–º –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å —Ç–∞–∫–∏–µ —Ç–≤–∏—Ç—ã –≤ —Å–ø–∏—Å–æ–∫.

```python
deleted_twits = []
for twit in twitter:
    for key in twit:
        if key == 'delete':
            deleted_twits.append(twit)
per_deleted = len(deleted_twits)/len(twitter)*100
print('–ø—Ä–æ—Ü–µ–Ω—Ç —É–¥–∞–ª–µ–Ω–Ω—ã—Ö —Ç–≤–∏—Ç–æ–≤ —Ä–∞–≤–µ–Ω', per_deleted)
```

# –í–æ–ø—Ä–æ—Å 3

–ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ —É —Ç–≤–∏—Ç–∞ –µ—Å—Ç—å —Ç—ç–≥ 'lang', –∑–∞–ø–∏—à–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–æ–≥–æ —Ç—ç–≥–∞ –≤ —Å–ø–∏—Å–æ–∫.

```python
languages = []

for twitt in twitter:
    language = twitt.get('lang')
    if language is not None:
        languages.append(language)

freq_languages = Counter(languages)
top_10_lang = freq_languages.most_common(10)

for data in top_10_lang:
    print(data[0], '-', data[1])
```

# –í–æ–ø—Ä–æ—Å 4

–ë—É–¥–µ–º –∏—Å–∫–∞—Ç—å id –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è, –ø–æ—Ç–æ–º—É —á—Ç–æ –∏–º–µ–Ω–∞ —É –¥–≤—É—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –º–æ–≥—É—Ç —Å–æ–≤–ø–∞–¥–∞—Ç—å


```python
names = []
for twitt in twitter:
    user = twitt.get('user')
    if user is not None:
        identifiction = user.get('id')
        names.append(identifiction)
```

```python
repeating_names = []
users = Counter(names)
for user in users:
    if users[user] > 1:
        repeating_names.append(user)
print('–¢–∞–∫–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –µ—Å—Ç—å, –∏—Ö', len(repeating_names))
```

# –í–æ–ø—Ä–æ—Å 5

–° –ø–æ–º–æ—â—å—é –∫–ª—é—á–µ–π 'entities', 'hashtags', 'text' –Ω–∞–π–¥–µ–º —Ö—ç—à—Ç–µ–≥–∏, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω—ã–µ –≤ —Ç–≤–∏—Ç–µ

```python
texts = []
for twitt in twitter:
    entities = twitt.get('entities')
    if entities is not None:
        hashtags = entities.get('hashtags')
        if hashtags is not None:
            for hashtag in hashtags:
                text = hashtag['text']
                texts.append(text)

freq_text = Counter(texts)

for hashtag in freq_text.most_common(20):
    print(hashtag[0], hashtag[1])
```

# –í–æ–ø—Ä–æ—Å 6

–ø—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ —è–∑—ã–∫ —Ç–≤–∏—Ç–∞ - –∞–Ω–≥–ª–∏–π—Å–∫–∏–π, —á—Ç–æ –æ–Ω –Ω–µ –±—ã–ª —Ä–µ—Ç–≤–∏—Ç–Ω—É—Ç, –∏ —á—Ç–æ –µ–≥–æ –Ω–µ —Ü–∏—Ç–∏—Ä–æ–≤–∞–ª–∏.

–ò–∑–±–∞–≤–∏–º—Å—è –æ—Ç–ø—É–Ω–∫—Ç—É–∞—Ü–∏–∏ –∏ –ø—Ä–∏–≤–µ–¥–µ–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É.


```python
import string

all_texts = []
for twitt in twitter:
    lang = twitt.get('lang')
    if lang == 'en':
        if twitt.get('retweeted_status') is None:
            if twitt.get('is_quote_status') is False:
                text = twitt.get('text').lower()
                text = text.translate(text.maketrans(
                    '', '', string.punctuation))
                all_texts.append(text)
```

–†–∞–∑–æ–±—å–µ–º –Ω–∞—à —Å–ø–∏—Å–æ–∫ –∏–∑ —Ç–≤–∏—Ç—Ç–æ–≤ –Ω–∞ —Å–ª–æ–≤–∞, –ø–æ—Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É.

```python
words = []
for text in all_texts:
    words.extend(text.split(' '))

Counter(words).most_common()
```

# –í–æ–ø—Ä–æ—Å 7 

–ù–∞–π–¥–µ–º –∏–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –µ–≥–æ –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤, —Å–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ –∏–∑ –∫–æ—Ä—Ç–µ–∂–µ–π —Ç–∏–ø–∞ (–∏–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è, –∫–æ–ª-–≤–æ –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤), –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é.

```python
from operator import itemgetter

data = []

for twitt in twitter:
    user = twitt.get('user')
    if user is not None:
        name = user.get('name')
        num_followers = user.get('followers_count')
        d = (name, num_followers)
        data.append(d)
        
sorted(data, key=itemgetter(1), reverse=True)[:10]
```

# –í–æ–ø—Ä–æ—Å 8

–ù–∞–π–¥–µ–º —Å—Ç—Ä–æ–∫–∏, —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ –∏—Å—Ç–æ—á–Ω–∏–∫, —Å –ø–æ–º–æ—â—å—é —Ä–µ–≥—É–ª—Ä—è–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –≤—ã—Ç–∞—â–∏–º –Ω—É–∂–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é.


```python
import re

list_sources = []

for twitt in twitter:
    source = twitt.get('source')
    if source is not None:
        sources = re.findall(r'>(.+[^<])<', source)
        list_sources.append(sources[0])

Counter(list_sources).most_common(10)
```
