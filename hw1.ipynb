```python
%load_ext pycodestyle_magic
%pycodestyle_on

import json
from pprint import pprint
from collections import Counter

twitter = []
for line in open('hw_3_twitter.json'):
    twitter.append(json.loads(line))
pprint(twitter[0])
```

# Вопрос 2

Проверим, является ли ключ твита строкой 'deleted'. Если да, будем записывать такие твиты в список.

```python
deleted_twits = []
for twit in twitter:
    for key in twit:
        if key == 'delete':
            deleted_twits.append(twit)
per_deleted = len(deleted_twits)/len(twitter)*100
print('процент удаленных твитов равен', per_deleted)
```

# Вопрос 3

Проверим, что у твита есть тэг 'lang', запишем значение этого тэга в список.

```python
languages = []

for twitt in twitter:
    language = twitt.get('lang')
    if language is not None:
        languages.append(language)

freq_languages = Counter(languages)
top_10_lang = freq_languages.most_common(10)

for data in top_10_lang:
    print(data[0], '-', data[1])
```

# Вопрос 4

Будем искать id пользователя, потому что имена у двух пользователей могут совпадать


```python
names = []
for twitt in twitter:
    user = twitt.get('user')
    if user is not None:
        identifiction = user.get('id')
        names.append(identifiction)
```

```python
repeating_names = []
users = Counter(names)
for user in users:
    if users[user] > 1:
        repeating_names.append(user)
print('Такие пользователи есть, их', len(repeating_names))
```

# Вопрос 5

С помощью ключей 'entities', 'hashtags', 'text' найдем хэштеги, использованные в твите

```python
texts = []
for twitt in twitter:
    entities = twitt.get('entities')
    if entities is not None:
        hashtags = entities.get('hashtags')
        if hashtags is not None:
            for hashtag in hashtags:
                text = hashtag['text']
                texts.append(text)

freq_text = Counter(texts)

for hashtag in freq_text.most_common(20):
    print(hashtag[0], hashtag[1])
```

# Вопрос 6

проверим, что язык твита - английский, что он не был ретвитнут, и что его не цитировали.

Избавимся отпунктуации и приведем к нижнему регистру.


```python
import string

all_texts = []
for twitt in twitter:
    lang = twitt.get('lang')
    if lang == 'en':
        if twitt.get('retweeted_status') is None:
            if twitt.get('is_quote_status') is False:
                text = twitt.get('text').lower()
                text = text.translate(text.maketrans(
                    '', '', string.punctuation))
                all_texts.append(text)
```

Разобьем наш список из твиттов на слова, посчитаем частоту.

```python
words = []
for text in all_texts:
    words.extend(text.split(' '))

Counter(words).most_common()
```

# Вопрос 7 

Найдем имя пользователя и количество его подписчиков, создадим список из кортежей типа (имя пользователя, кол-во подписчиков), отсортируем по убыванию.

```python
from operator import itemgetter

data = []

for twitt in twitter:
    user = twitt.get('user')
    if user is not None:
        name = user.get('name')
        num_followers = user.get('followers_count')
        d = (name, num_followers)
        data.append(d)
        
sorted(data, key=itemgetter(1), reverse=True)[:10]
```

# Вопрос 8

Найдем строки, содержащие источник, с помощью регулряного выражения вытащим нужную информацию.


```python
import re

list_sources = []

for twitt in twitter:
    source = twitt.get('source')
    if source is not None:
        sources = re.findall(r'>(.+[^<])<', source)
        list_sources.append(sources[0])

Counter(list_sources).most_common(10)
```
